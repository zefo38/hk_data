```python
import seaborn as sns 

tips = sns.load_dataset("tips")
tips.head()
```

## GroupBy
- 데이터셋.groupby(컬럼명).집계함수
-       해당 컬럼은 범주 형태
- 포인트 1. 데이터 형태가 바뀜
-       2. 추가적인 조작

```python
tips.head()
```

```python
tips.groupby('sex',observed=True)['tip'].mean()
```

```python
tips.groupby('sex',observed=True)['tip'].mean().reset_index()
```

```python
tips.groupby(['sex','day'],observed=True)['tip'].mean().reset_index()
```

## 문제
- total_bill이 20보다 큰 행들만 필터링하세요.

```python
# 코드
tips.loc[tips['total_bill'] > 20]
```

- total_bill이 20보다 크고 tip이 3보다 큰 행들만 필터링하세요.

```python
# 코드
a = tips['total_bill'] > 20
b = tips['tip'] > 3
tips.loc[ a & b,:].reset_index(drop=True)
```

## 문제
- New가 들어간 도시만 출력
- contains 메서드 활용

```python
import pandas as pd

data = {
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'city': ['New York', 'Los Angeles', 'New Yorker', 'Chicago', 'New City'],
    'age': [25, 30, 35, 40, 45]
}

df = pd.DataFrame(data)
df
```

```python
# code
df.loc[df['city'].str.contains('New')]
```
