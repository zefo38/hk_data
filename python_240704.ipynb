## 시각화 연습
- matplotlib와 seaborn을 활용하여 다양한 시각화를 해본다.
- 시각화는 2 x 2 형태로 나오도록 한다. (예시)
![](tips_visualization.png)
- 시각화는 seaborn 라이브러리 참조해서 적용 (~9시 20분)

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
```

```python
tips.info()
```

```python
import warnings

# 경고 메시지 무시
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic' # 옵션 설정
fig, ax = plt.subplots(nrows = 2, ncols=2 , figsize=(14, 10))

ax1 = sns.scatterplot(x = 'total_bill', y= 'tip', data=tips,
                      hue = 'time', ax= ax[0,0]
                     )
ax1.set_title("총 금액과 팁")

ax2 = sns.boxplot(x = 'day', y ='total_bill', data= tips,
                    hue = 'sex', ax=ax[0,1])
ax2.set_title("요일 및 성별에 따른 팁")


ax3 = sns.violinplot(x='day',y='tip', data= tips,
                    hue = 'smoker', ax=ax[1,0])
ax3.set_title("요일 및 흡연 여부에 따른 팁")

ax4 = sns.histplot(x='total_bill', data= tips
                   , ax=ax[1,1])
ax4.set_title("총 금액 히스토그램")

plt.tight_layout() # 간격 조정
plt.savefig('yourname.png', dpi=300) # 시각화 내보내기
plt.show()
```
