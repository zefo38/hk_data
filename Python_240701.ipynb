```python
import os
import pandas as pd

def read_csv_files_from_directory(directory_path):
    # List all files in the directory
    files = os.listdir(directory_path)
    
    # Filter out only the CSV files
    csv_files = [file for file in files if file.endswith('.csv')]
    
    # Dictionary to hold DataFrames
    dataframes = {}
    
    # Loop over the list of CSV files and read each into a DataFrame
    for csv_file in csv_files:
        file_path = os.path.join(directory_path, csv_file)
        df = pd.read_csv(file_path)
        # Store the DataFrame in the dictionary with the file name (without extension) as the key
        file_name = os.path.splitext(csv_file)[0]
        dataframes[file_name] = df
        print(f"Loaded {csv_file} into DataFrame")
    
    return dataframes

# Example usage
directory_path = './kaggle'
dataframes = read_csv_files_from_directory(directory_path)

# Process each DataFrame as needed
for name, df in dataframes.items():
    print(f"DataFrame {name}:")
    print(df.head())  # Print the first few rows of each DataFrame
```

```python
import os
import pandas as pd

def read_csv_files_from_directory(directory_path):
    # List all files in the directory
    files = os.listdir(directory_path)
    
    # Filter out only the CSV files
    csv_files = [file for file in files if file.endswith('.csv')]
    
    # Loop over the list of CSV files and read each into a DataFrame
    for csv_file in csv_files:
        file_path = os.path.join(directory_path, csv_file)
        df = pd.read_csv(file_path)
        
        # Create a variable name based on the file name (without extension)
        file_name = os.path.splitext(csv_file)[0]
        
        # Create a global variable with the DataFrame
        globals()[file_name] = df
        
        print(f"Loaded {csv_file} into DataFrame '{file_name}'")
    
# Example usage
directory_path = './kaggle'
read_csv_files_from_directory(directory_path)

# Now each DataFrame is available as a global variable
# Example: if there's a file named 'data1.csv', the DataFrame would be named 'data1'
```
